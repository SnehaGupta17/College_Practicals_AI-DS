# Data Visualization I

## 1.Use the inbuilt dataset 'titanic'. The dataset contains 891 rows and contains information about the passengers who boarded the unfortunate Titanic ship. Use the Seaborn library to see if we can find any patterns in the data. 

#### Step 1:Import Required Libraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

#### Step 2:Load dataset

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

#### Step 3: Distribution Plots

#### A.Histogram of Ticket Fare

```python
plt.figure(figsize=(8, 6))
sns.histplot(dataset['fare'], bins=30, kde=True)
plt.title("Ticket Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

#### B.Distplot for Passenger Age

```python
plt.figure(figsize=(8, 6))
sns.histplot(dataset['age'], bins=10, kde=True)
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

#### C.Joint Plot: Age vs. Fare (Scatter)

```python
sns.jointplot(x=dataset['age'], y=dataset['fare'], kind='scatter')
plt.show()
```

#### D.Joint Plot: Age vs. Fare (Hex)

```python
sns.jointplot(x=dataset['age'], y=dataset['fare'], kind='hex')
plt.show()
```

#### E.Rug Plot for Fare

```python
plt.figure(figsize=(8, 3))
sns.rugplot(dataset['fare'])
plt.title("Rug Plot for Fare")
plt.show()
```

#### Step 4: Categorical Plots

#### A.Bar Plot: Average Age of Male vs. Female Passengers

```python
plt.figure(figsize=(6, 4))
sns.barplot(x='sex', y='age', data=dataset)
plt.title("Average Age of Male and Female Passengers")
plt.show()
```

#### B.Count Plot: Number of Male vs. Female Passengers

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='sex', data=dataset)
plt.title("Number of Male and Female Passengers")
plt.show()
```

#### C.Box Plot: Age Distribution by Gender

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='sex', y='age', data=dataset)
plt.title("Box Plot of Age by Gender")
plt.show()
```

#### D.Box Plot with Hue: Age Distribution by Gender and Survival

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='sex', y='age', data=dataset, hue='survived')
plt.title("Box Plot of Age by Gender and Survival")
plt.show()
```

#### E.Violin Plot: Age Distribution by Gender

```python
plt.figure(figsize=(6, 4))
sns.violinplot(x='sex', y='age', data=dataset)
plt.title("Violin Plot of Age by Gender")
plt.show()
```

#### Step 5: Advanced Plots
#### Strip Plot: Age vs. Gender

```python
plt.figure(figsize=(6, 4))
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
plt.title("Strip Plot of Age by Gender")
plt.show()
```

#### Swarm Plot: Age vs. Gender

```python
plt.figure(figsize=(6, 4))
sns.swarmplot(x='sex', y='age', data=dataset)
plt.title("Swarm Plot of Age by Gender")
plt.show()
```

#### Step 6: Matrix Plots
#### Heat Map of Correlation Matrix

#### plt.figure(figsize=(10, 6))

#### Convert categorical variables into numeric using one-hot encoding

```python
dataset_numeric = dataset.select_dtypes(include=['number'])  # Select only numeric columns

corr = dataset_numeric.corr()  # Compute correlation on numerical data
sns.heatmap(corr, annot=True, cmap="magma")

plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```

#### Cluster Map

```python
sns.clustermap(corr, cmap="magma", annot=True)
plt.title("Cluster Map of Titanic Dataset")
plt.show()
```
