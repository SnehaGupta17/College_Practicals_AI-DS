# Data Visualization II

## 1.Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age')

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='sex', y='age', hue='survived', palette='Set1')

# Set the title and labels
plt.title('Distribution of Age by Gender and Survival Status', fontsize=14)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Age', fontsize=12)

# Show the plot
plt.show()
```

## 2.Write observations on the inference from the above statistics.

```python
Observations and Inferences from the Box Plot (Based on the Code):
Age Distribution by Gender:

Males (sex = male):

The age distribution for male passengers shows a wide range, with the box plot indicating a broad spread of ages.

The median age for males is slightly higher than for females, suggesting that there were more adult and older male passengers on board.

There are several outliers, especially among non-surviving males, indicating that a few older individuals didn't survive the disaster.

Females (sex = female):

The age distribution for female passengers is more concentrated around the middle, with a narrower range compared to males.

The median age for females is lower than for males, indicating that the majority of female passengers were younger.

Outliers are fewer in the female category, particularly in the non-survivors group.

Age Distribution Based on Survival Status:

Survivors (survived = 1):

Both male and female survivors have younger ages, with the female survivors showing a very clear concentration of younger individuals.

The median age of survivors is lower for both genders, with females particularly clustered around a younger age.

For males, there is a slight concentration of survivors in the younger age range, but the spread is still wide compared to females.

Non-survivors (survived = 0):

The non-survivors (both males and females) show a higher median age, with a wider spread of ages, particularly among the males.

The age range for non-surviving males is much broader, indicating that older males had a lower chance of survival compared to younger males.

Female non-survivors also have a slightly higher median age, but the spread is narrower compared to males. More young females survived than non-survived.

Outliers:

The box plot indicates outliers for both male and female non-survivors. For example, several older male passengers (in their 50s and 60s) are marked as outliers, suggesting they did not survive the disaster.

Outliers among female non-survivors are less prominent, but there are still a few older individuals, though they are fewer in number.

Key Trends and Inferences:

The code generates a box plot that visually shows that younger passengers (both male and female) had a higher likelihood of survival.

Females show a much more concentrated age distribution, with a higher number of young female survivors, indicating that women and children were likely prioritized during the Titanic's evacuation.

Males, particularly older males, had a higher median age among non-survivors, suggesting that older male passengers were less likely to survive the disaster.

The narrower spread of age for females (especially survivors) implies a relatively younger demographic compared to males on the Titanic.

Conclusion:
The box plot created by the code visually demonstrates that young women and children had a higher chance of survival, which aligns with historical accounts of the Titanic disaster. Older males, on the other hand, had a lower chance of survival, with many older males being in the non-survivor group.
```
