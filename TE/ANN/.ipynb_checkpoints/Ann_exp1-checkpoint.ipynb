#### Exp 1 : Plotting few activation functions used in neural networks

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sig(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10,10,50)
p=sig(x)
plt.plot(x,p)
plt.title("Sigmoid Activation Function")
plt.show()
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-10,10)
p=sig(x)
plt.plot(x,tanh(x))
plt.title("TanH Activation Function")
plt.show()
```

```python

# Activation Functions
def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))  # stability fix
    return e_x / np.sum(e_x)
# ReLU Plot
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.plot(x, relu(x), label="ReLU", color='orange')
plt.title("ReLU Activation Function")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

```

```python
# Softmax Plot (over multiple inputs)
softmax_outputs = np.array([softmax(np.array([xi, 0, -xi]))[0] for xi in x])
plt.subplot(1, 2, 2)
plt.plot(x, softmax_outputs, label="Softmax (x vs softmax([x, 0, -x]))", color='blue')
plt.title("Softmax (single component)")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
