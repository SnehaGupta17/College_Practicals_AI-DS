# Write a program for performing industrial data analysis using relevant tools and techniques

```python
import pandas as pd
import matplotlib.pyplot as plt
import re

# Read Excel file (change the path as per your system)
df = pd.read_excel("Copy of arduino_nano_uno.xlsx", header=None)     #Modified and corrected og file, replaced 12 am to 12 pm

# Add column names
df.columns = ['Time', 'Farm1_Temp', 'Farm1_Humidity', 'Farm2_Temp', 'Farm2_Humidity',
              'Farm3_Temp', 'Farm3_Humidity', 'Farm4_Temp', 'Farm4_Humidity']
df.head()
```

```python
# --- Extract Hour Manually from Text like "0:00:00 AM", "1:00:00 PM" ---
def extract_hour(time_str):
    match = re.match(r"(\d+):\d+:\d+ (\w+)", str(time_str).strip())
    if match:
        hour = int(match.group(1))
        am_pm = match.group(2).upper()
        # Convert 12-hour to 24-hour
        if am_pm == "PM" and hour != 12:
            hour += 12
        elif am_pm == "AM" and hour == 12:
            hour = 0
        return hour
    return None

df["Hour"] = df["Time"].apply(extract_hour)
```

```python
# --- Slice dataset per day (24 readings each) ---
rows_per_day = 24
num_days = len(df) // rows_per_day

def get_day_data(day_index):
    start = day_index * rows_per_day
    end = (day_index + 1) * rows_per_day
    return df.iloc[start:end]
```

```python
# -------------------------------------
# Scatter Plot – 1 Day Temperature
# -------------------------------------
day1 = get_day_data(0)
plt.figure(figsize=(8,4))
plt.scatter(day1['Hour'], day1['Farm1_Temp'], color='red')
plt.title('Scatter Plot – Farm 1 Temperature (Day 1)')
plt.xlabel('Time (Hour)')
plt.ylabel('Temperature (°C)')
plt.xticks(range(0,24), [f"{h%12 or 12} {'AM' if h<12 else 'PM'}" for h in range(24)], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------
# Line Plot – 2 Days Temperature
# -------------------------------------
plt.figure(figsize=(10,4))
colors = ['red', 'blue']
for i in range(2):
    day_data = get_day_data(i)
    plt.plot(day_data['Hour'], day_data['Farm1_Temp'], marker='o',
             color=colors[i], label=f'Day {i+1} Temp')
plt.title('Line Plot – Farm 1 Temperature (2 Days)')
plt.xlabel('Time (Hour)')
plt.ylabel('Temperature (°C)')
plt.xticks(range(0,24), [f"{h%12 or 12} {'AM' if h<12 else 'PM'}" for h in range(24)], rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------
# Clustered Bar Graph – 1 Day Humidity
# -------------------------------------
plt.figure(figsize=(8,4))
plt.bar(day1['Hour'], day1['Farm1_Humidity'], width=0.4, color='skyblue', edgecolor='black')
plt.title('Clustered Bar Graph – Farm 1 Humidity (Day 1)')
plt.xlabel('Time (Hour)')
plt.ylabel('Humidity (%)')
plt.xticks(range(0,24), [f"{h%12 or 12} {'AM' if h<12 else 'PM'}" for h in range(24)], rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------
# Scatter Plot – 1 Day Humidity
# -------------------------------------
plt.figure(figsize=(8,4))
plt.scatter(day1['Hour'], day1['Farm1_Humidity'], color='purple')
plt.title('Scatter Plot – Farm 1 Humidity (Day 1)')
plt.xlabel('Time (Hour)')
plt.ylabel('Humidity (%)')
plt.xticks(range(0,24), [f"{h%12 or 12} {'AM' if h<12 else 'PM'}" for h in range(24)], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------
# Line Plot – 2 Days Humidity
# -------------------------------------
plt.figure(figsize=(10,4))
colors = ['green', 'orange']
for i in range(2):
    day_data = get_day_data(i)
    plt.plot(day_data['Hour'], day_data['Farm1_Humidity'],
             marker='s', color=colors[i], label=f'Day {i+1} Humidity')
plt.title('Line Plot – Farm 1 Humidity (2 Days)')
plt.xlabel('Time (Hour)')
plt.ylabel('Humidity (%)')
plt.xticks(range(0,24), [f"{h%12 or 12} {'AM' if h<12 else 'PM'}" for h in range(24)], rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
